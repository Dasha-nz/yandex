```python
# чтение файла и просмотр загруженных данных
df = pd.read_csv('gdrive/My Drive/L.2.csv')
# df = pd.read_excel('L.2.xlsx')
df.head()
```

```python
#удаление столбцов
df = df.drop(['TY'], axis=1)
df = df.drop(['JO'], axis=1)
df = df.drop(['VL'], axis=1)
df = df.drop(['SP'], axis=1)
df = df.drop(['DA'], axis=1)
df = df.drop(['SN'], axis=1)
df = df.drop(['UR'], axis=1)
df = df.drop(['EP'], axis=1)
df = df.drop(['IS'], axis=1)
df = df.drop(['T2'], axis=1)
df = df.drop(['We'], axis=1)
df = df.drop(['In'], axis=1)
df.head(10)
```

```python
#Преобразование столбца AU в строку
df['AU'] = df['AU'].astype(str)
```

```python
#Функция swap_words разбивает на слова возвращает новую строку с замененными словами.apply применяет это ко всему столбцу
def swap_words(text):
    words = text.split()
    if len(words) > 1:
        words = words[::-1]  # reverse the list of words
    return ' '.join(words)
df['AU'] = df['AU'].apply(swap_words)
df.head(10)
```

```python
#Переименовала столбцы
df.rename(columns = {'T1':'Title',
                     'AU':'Author',
                     'PY':'Year', 'DO':'DOI',
                     'KW':'Key Words',
                     'AB':'Abstract'},
          inplace = True)
df.head()
```

```python
#Поменяла местами столбцы
df = df[['Title', 'Author', 'Year', 'Key Words', 'DOI', 'Abstract']]
df.head(6)
```

```python
df.fillna(method='ffill', inplace=True)
df.head(5)
```

```python
#Объединенила в список
df = df.groupby(['Title', 'Year', 'DOI', 'Abstract'], as_index=True).agg({'Author': lambda x: list(x), 'Key Words': lambda x: list(x)})
df.head(2)
```

```python
df = df.reset_index()
df.head(6)
```

```python
#Поменяла местами столбцы
df = df[['Title', 'Author', 'Year', 'Key Words', 'DOI', 'Abstract']]
df.head(6)
```
